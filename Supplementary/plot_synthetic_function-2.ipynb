```python
import numpy as np
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
```

```python
x0 = np.linspace(0, 1, 100)
x1 = np.linspace(0, 10, 100)
X0, X1 = np.meshgrid(x0, x1)
```

```python
epsilon = np.random.normal(0, 1, size=X0.shape)
```

```python
Y = np.sin(2 * np.pi * X0) + np.log10(X1 + 1) + 0.5 * epsilon
```

```python
# Plotting
fig = plt.figure(figsize=(12, 7))
ax = fig.add_subplot(111, projection="3d")
surf = ax.plot_surface(X0, X1, Y, cmap="viridis", edgecolor="none")

ax.set_title("3D Plot of y = sin(2œÄX‚ÇÄ) + log‚ÇÅ‚ÇÄ(X‚ÇÅ + 1) + 0.5Œµ")
ax.set_xlabel("X‚ÇÄ")
ax.set_ylabel("X‚ÇÅ")
ax.set_zlabel("y")

fig.colorbar(surf, shrink=0.5, aspect=10)

plt.tight_layout()
plt.show()
```
